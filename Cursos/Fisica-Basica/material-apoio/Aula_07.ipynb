### Carga Elétrica

### A Força Elétrica - Lei de Coulomb

O módulo da força elétrica entre duas cargas puntiformes é diretamente proporcional ao produto das cargas e inversamente proporcional ao quadrado da distância entre elas.

Módulo da Força Elétrica
$$F = k \frac{q_{1}q_{2}}{r^{2}}$$

$$\vec{F} = k\frac{q_{1}q_{2}}{r^{2}}\hat{r}$$

### Campo Elétrico

O Campo Elétrico é a propriedade das cargas elétricas de influenciarem sua vizinhança.

A força elétrica sobre um corpo carregado é exercida pelo campo elétrico
produzido por outros corpos carregados.

$$\vec{E} = \frac{\vec{F}}{q_{0}}$$

$$\vec{E} = kq\frac{\hat{r}}{r^{2}} $$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Criando o Conjunto de dados do campo
w = 3
Y, X = np.mgrid[-w:w:100j, -w:w:100j] 
# Campo elétrico da Carga puntiforme
U = +X/(X**2+Y**2)**(3/2)
V = +Y/(X**2+Y**2)**(3/2)
# Creating plot 
fig, ax = plt.subplots(figsize = (8, 8)) 
# Pontos representando as cargas
ax.scatter(0,0, color='r')
# Gráfico de campo elétrico
ax = plt.streamplot(X, Y, U, V, density = 3, color='g') 
 # show plot 
plt.show() 
```

### Energia Potencial Elétrica

Primeiro deve-se determinar a energia potencial elétrica entre duas cargas
$$W = \int\limits_{r_{a}}^{r_{b}}\vec{F} \cdot d\vec{r}$$

$$W = \int\limits_{r_{a}}^{r_{b}}F dl\cos \phi$$

<img src='imgs/PotencialEletrico.jpg' width='400'/>

$$W = \int\limits_{r_{a}}^{r_{b}}\frac{1}{4 \pi \varepsilon_{0}}\frac{qq_{0}}{r^{2}}dr$$

$$W = - \Delta U = - (U_{b} - U_{a}) = \int\limits_{r_{a}}^{r_{b}}\frac{1}{4 \pi \varepsilon_{0}}\frac{qq_{0}}{r^{2}}dr $$

$$-\Delta U = -(U_{b} - U_{a}) = \frac{qq_{0}}{4 \pi \varepsilon_{0}}\left(-\frac{1}{r}\right)_{r_{a}}^{r_{b}}$$

$$-\Delta U = -(U_{b} - U_{a}) = -\frac{qq_{0}}{4 \pi \varepsilon_{0}}\left(\frac{1}{r_{b}} - \frac{1}{r_{a}} \right)$$

$$U = \frac{qq_{0}}{4\pi\varepsilon_{0}}\frac{1}{r}$$

A energia potencial é sempre definida em relação a algum ponto no qual $U = 0$. U é igual a zero quando a distância entre $q$ e $q_{0}$ é muito grande e $r \to \infty$. Portanto, $U$ é igual ao trabalho realizado pelo campo elétrico de q para deslocar $q_{0}$ de uma distância inicial $r$ até o infinito. Quando $q$ e $q_{0}$ possuem o mesmo sinal, a interação é repulsiva e o trabalho é positivo, de modo que $U$ é positiva para qualquer distância finita. Quando as cargas possuem sinais opostos, a interação é atrativa e o trabalho é negativo, de modo que $U$ é negativa.

### Potencial Elétrico

Denomina-se potencial elétrico a energia potencial por unidade de carga.

$$V = \frac{U}{q_{0}}$$

$$V = \frac{1}{4\pi\varepsilon_{0}}\frac{q}{r}$$

Dipolo Elétrico

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Criando o Conjunto de dados do campo
w = 5
Y, X = np.mgrid[-w:w:100j, -w:w:100j] 
# Função que descreve o comportamento do dipolo elétrico
z = (1/np.sqrt((X+0.5)**2+Y**2)+1/np.sqrt((X-0.5)**2+Y**2)+1/np.sqrt((X+1.5)**2+Y**2)
     +1/np.sqrt((X-1.5)**2+Y**2)+1/np.sqrt((X-2.5)**2+Y**2)+1/np.sqrt((X+2.5)**2+Y**2)
    +1/np.sqrt((X-3.5)**2+Y**2)+1/np.sqrt((X+3.5)**2+Y**2))
# Valores das coordenadas do campo elétrico
V,U = np.gradient(-z)

# Creating plot 
fig, ax = plt.subplots(figsize = (8, 8)) 
# Pontos representando as cargas
ax.scatter(-0.5,0, color='b')
ax.scatter(+0.5,0, color='b')
ax.scatter(+1.5,0, color='b')
ax.scatter(-1.5,0, color='b')
ax.scatter(+2.5,0, color='b')
ax.scatter(-2.5,0, color='b')
ax.scatter(+3.5,0, color='b')
ax.scatter(-3.5,0, color='b')
# Gráfico de campo elétrico
ax = plt.streamplot(X, Y, U, V, density = 3, color='g') 
 # show plot 
plt.show() 
```
