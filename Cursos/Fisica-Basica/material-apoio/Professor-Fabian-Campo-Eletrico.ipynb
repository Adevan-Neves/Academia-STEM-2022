```python
 import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# Adevan Neves Santos Engenharia de Computação
 
# Criando o Conjunto de dados do campo
w = 5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
# Função que descreve o comportamento do dipolo elétrico
z = (1/np.sqrt((X+0.5)**2+Y**2)+1/np.sqrt((X-0.5)**2+Y**2)-2/np.sqrt((X)**2+(Y**2)))
 
# Valores das coordenadas do campo elétrico
V,U = np.gradient(-z)
total=z.sum()
print("O potencial eletrico total foi de : "+str(total)+" V")
 
 
# Creating plot 
fig, ax = plt.subplots(figsize = (8, 8))
# Pontos representando as cargas
ax.scatter(-0.5,0, color='b')
ax.scatter(+0.5,0, color='b')
ax.scatter(0,0, color='r')
 
# Gráfico de campo elétrico
ax = plt.streamplot(X, Y, U, V, density = 3, color='g')
# show plot 
plt.show()
plt.scatter(V,U)
plt.show()
```
